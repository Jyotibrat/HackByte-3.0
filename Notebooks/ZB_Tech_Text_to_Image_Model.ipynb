```python
import torch
from diffusers import DiffusionPipeline

# Model name from Hugging Face
model_name = "ZB-Tech/Text-to-Image"

# Load the pipeline
pipe = DiffusionPipeline.from_pretrained(model_name, torch_dtype=torch.float16)

# Move to GPU if available
device = "cuda" if torch.cuda.is_available() else "cpu"
pipe.to(device)
```

```python
def generate_image(prompt):
    print(f"Generating image for prompt: {prompt}")
    image = pipe(prompt).images[0]  # Generate image
    return image
```

```python
from IPython.display import display
from PIL import Image

# Example prompt
user_prompt = "Generate a floor plan of a 3 bedroom apartment with all the rooms labeled in english."
image = generate_image(user_prompt)

# Display the generated image
image.show()

image.save("generated_image.png")

img = Image.open("generated_image.png")
display(img)
```
